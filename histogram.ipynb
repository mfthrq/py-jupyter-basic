```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from turtle import color
```

```python
subjects = ["Math", "English", "Art", "Chemistry", "History", "Tech", "Music"]
n_stud = [2,3,5,4,3,6,4]
```

```python
plt.figure(figsize=(10,5))
plt.bar(subjects, n_stud, color='r')
plt.title("Fav subj")
plt.show()
```

```python
school = ['SD', 'SMP', 'SMA']
green = [4,2,1]
blue = [2,5,4]
yellow = [3,6,5]
red = [1,4,2]

df = pd.DataFrame([school, green, blue, yellow, red]).T
df.columns = ['school', 'green', 'blue', 'yellow', 'red']
df
```

```python
colors = ['g', 'b', 'y', 'r']
df.plot(x = 'school',
        kind = 'bar',
        color = colors,
        rot = 0,
        title = 'Fav color of studs'
)

plt.show()
```

```python
# Data 500 orang

age = np.random.gamma(2,4,500).astype(int)
age[:10]
```

```python
plt.hist(age, bins=10)
plt.xlabel('Umur')
plt.ylabel('Jumlah')
```

```python
sns.histplot(age, bins=10)
plt.xlabel('Umur')
plt.ylabel('Jumlah Penduduk')
```
