Descriptive Statistic

```python
import statistics as stat
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gaji = np.random.normal(5000, 100, size=100000).astype(int)
gaji
```

```python
print('Modus = ', stat.mode(gaji))
```

```python
from turtle import color


sns.histplot(gaji, bins=20)
plt.axvline(stat.mode(gaji), color='y')
plt.xlabel("GAJI")
```

```python
umur = np.random.gamma(4,5,200).astype(int)
umur
```

```python
print("Mean umur : ", np.mean(umur))
```

```python
sns.histplot(umur, bins=20)
plt.title("MEAN, MODUS, MEDIAN PADA 200 UMUR")
plt.axvline(np.mean(umur), color='r', linestyle="--", label="MEAN")
plt.axvline(stat.mode(umur), color='y', linestyle="--", label="MODE")
plt.axvline(stat.median(umur), color='g', linestyle="--", label="MEDIAN")
plt.legend()
plt.xlabel("UMUR")
```
