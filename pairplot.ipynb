```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.pairplot(
    data = sns.load_dataset("flights")
)
```

```python
dataset = sns.load_dataset("flights").copy()
```
